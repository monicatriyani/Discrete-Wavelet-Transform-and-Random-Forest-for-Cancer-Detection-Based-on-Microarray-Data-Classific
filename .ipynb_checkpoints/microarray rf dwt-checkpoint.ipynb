### Normalisasi

```python
def standard_scaler(X_train_, X_test_):
    scaler = StandardScaler()
    X_train_ = scaler.fit_transform(X_train_)
    X_test_ = scaler.transform(X_test_)
    return X_train_,X_test_
```

```python
def minmax_scaler(X_train_, X_test_):
    scaler = MinMaxScaler()
    X_train_ = scaler.fit_transform(X_train_)
    X_test_ = scaler.transform(X_test_)
    return X_train_,X_test_
```

### DWT

```python
def dwt(X_train, X_test):
    train_coef = pywt.dwt(X_train, 'db4')
    #(cA, cD) = train_coef
    test_coef = pywt.dwt(X_test, 'db4')
    #(cA, cD) = 
    return train_coef, test_coef
```

```python
def koef_train(train_coef):
    (cA, cD) = train_coef
    return (cA, cD)

def koef_test(test_coef):
    (cA, cD) = test_coef
    return (cA, cD)
```

### Classification

```python
model_rfg = RandomForestClassifier(criterion='gini', n_estimators=60)
```

```python
X_colon_ = X_colon.values
y_colon_ = y_colon.values

X_lung_ = X_lung.values
y_lung_ = y_lung.values

X_cns_ = X_cns.values
y_cns_ = y_cns.values

X_breast_ = X_breast.values
y_breast_ = y_breast.values

X_ovarian_ = X_ovarian.values
y_ovarian_ = y_ovarian.values

X_prostate_ = X_prostate.values
y_prostate_ = y_prostate.values
```

**K-fold Cross Validation**

```python
kf_colon = KFold(n_splits=8, random_state=20)
kf_colon.get_n_splits(X_colon_)
```

```python
kf_lung = KFold(n_splits=8, random_state=20)
kf_lung.get_n_splits(X_lung_)
```

```python
kf_cns = KFold(n_splits=8, random_state=20)
kf_cns.get_n_splits(X_cns_)
```

```python
kf_breast = KFold(n_splits=8, random_state=20)
kf_breast.get_n_splits(X_breast_)
```

```python
kf_ovarian = KFold(n_splits=8, random_state=20)
kf_ovarian.get_n_splits(X_ovarian_)
```

```python
kf_prostate = KFold(n_splits=8, random_state=20)
kf_prostate.get_n_splits(X_prostate_)
```

**Random Forest with DWT**

- colon cancer

```python
print(kf_colon)
scoring = []
for train_index, test_index in kf_colon.split(X_colon_):
    X_train, X_test = X_colon_[train_index], X_colon_[test_index]
    y_train, y_test = y_colon_[train_index], y_colon_[test_index]
    X_train_, y_test_ = standard_scaler(X_train, X_test)
    train_coef, test_coef = dwt(X_train, X_test)
    (train_cA, train_cD) = koef_train(train_coef)
    (test_cA, test_cD) = koef_test(test_coef)
    model_rfg.fit(train_cA,y_train)
    pred = model_rfg.predict(test_cA)
    hasil = accuracy_score(y_test, pred)
    print("Score in Random Forest : ", hasil)
    scoring.append(hasil)
scoring_colon_std = np.array(scoring)
avg_rfg_std_colon = scoring_colon_std.mean()
print("avg accuraccy colon cancer = ",avg_rfg_std_colon)
print("max accuraccy colon cancer = ",scoring_colon_std.max())
print("min accuraccy colon cancer = ",scoring_colon_std.min())
```
